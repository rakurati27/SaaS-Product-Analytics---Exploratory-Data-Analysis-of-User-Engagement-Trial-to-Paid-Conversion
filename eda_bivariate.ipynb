# Bivariate Exploratory Data Analysis

This notebook compares user behavior against the conversion outcome (target).  
The objective is to identify which metrics differentiate **converted (1)** users from **non-converted (0)** users and drive performance above the **29% baseline conversion rate**.

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/Sandeep/Downloads/archive (1)/user_behavior_dataset.csv")
```

```python
# Analysis - 1 Visits vs Conversion

sns.boxplot(x="target", y="visits", data=df)
plt.title("Visits vs Conversion")
plt.show()

df.groupby("target")["visits"].median()
```

## Visits vs Conversion

The median number of visits is **9 for both converted and non-converted users**, showing **no meaningful difference** between the two groups.  
This indicates that visit frequency alone does **not** drive conversion and that returning to the platform without deeper engagement is insufficient to influence purchase behavior.


```python
# Analysis - 2 Clicks vs Conversion

sns.boxplot(x="target", y="clicks", data=df)
plt.title("Clicks vs Conversion")
plt.show()

df.groupby("target")["clicks"].median()
```

## Clicks vs Conversion

Converted users show a higher median of **6 clicks** compared to **5 clicks** for non-converted users, representing a **20% increase in interaction intensity**.  
This suggests that users who engage more actively with platform features are significantly more likely to convert, making clicks a key behavioral driver.


```python
# Analysis - 3 Time Spent vs Conversion

sns.boxplot(x="target", y="time_spent", data=df)
plt.title("Time Spent vs Conversion")
plt.show()

df.groupby("target")["time_spent"].median()
```

## Time Spent vs Conversion

Converted users spend a median of **369.5 seconds**, compared to **348 seconds** for non-converted users, which is an increase of approximately **6% more session time**.  
This confirms that deeper engagement duration is positively associated with conversion likelihood.


```python
# Analysis - 4 Correlation

plt.figure(figsize=(7,5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation with Target")
plt.show()
```

## Correlation Insights

The correlation matrix shows that **clicks and time_spent have the strongest positive association with the target variable**, while visits display minimal relationship.  
This reinforces that **quality of engagement** is more important than simple visit frequency in driving trial-to-paid conversion.


## Bivariate EDA Summary

- Visit count shows **no differentiation** between converters and non-converters  
- Converted users record **20% more clicks**, highlighting interaction depth as a key driver  
- Converted users spend **~6% more time** on the platform  
- Engagement quality (clicks, time_spent) matters more than frequency  
- Product efforts should focus on increasing in-app actions and session depth
